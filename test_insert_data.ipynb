```python
import weaviate
from weaviate.classes.init import Auth
import os
from dotenv import load_dotenv

load_dotenv()

WEAVIATE_URL = os.getenv("WEAVIATE_URL")
WCD_API_KEY = os.getenv("WEAVIATE_API_KEY")
OPENAI_API_KEY = os.getenv("OPENAI_API_KEY")

client = weaviate.connect_to_weaviate_cloud(
    cluster_url=WEAVIATE_URL,
    auth_credentials=Auth.api_key(WCD_API_KEY),
    headers={"X-OpenAI-Api-Key": OPENAI_API_KEY}
            )

if not client.is_ready():
  raise ConnectionError("Could not connect to Weaviate")
```

```python
collections = client.collections.list_all()
print("Collections:")
for col in collections:
    print("-", col)   # col is already a string
```

```python
col = client.collections.get("POI")
```

```python
for col in collections:
    collection = client.collections.get(col)
    print(f"\n--- {col} ---")

    # Get the first 10 objects
    results = collection.query.fetch_objects(limit=10)

    for obj in results.objects:
        print(obj.uuid, obj.properties)
```
